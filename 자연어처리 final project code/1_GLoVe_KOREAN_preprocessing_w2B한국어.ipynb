```python
from konlpy.tag import Okt, Komoran, Mecab, Hannanum, Kkma

def get_tokenizer(tokenizer=None): ## default : Okt
    if tokenizer == 'komoran':
        token = Komoran()
    elif tokenizer == 'kkma':
        token = Kkma()
    elif tokenizer == 'hannanum':
        token = Hannanum()
    elif tokenizer == 'okt':
        token = Okt()
    else:
        token = Mecab()
    return token
```

```python
tokenizer = get_tokenizer()
sentences2=sentences.apply(tokenizer.pos)
sentences2
```

#### 형태소 거름망

형태소 중 의미와 관련 있는 N(체언부), V(용언부), M(수식언부) 만 살림.  
J(조사)나 E(어미), X(접사)는 문법적 요소만을 위한 부분이므로 삭제

```python
def checkPOS(word):
    POS = ('NNG', 'NNP', 'NNB', 'NR', 'NP', 'VV', 'VA', 'VX', 'VCP', 'VCN', 'MM', 'MAG', 'MAJ', 'IC', 'XR')
    if word[1] in POS:
        return True
    return False
```

```python
#sentences3 = [word for sentence in sentences2 for word in sentence if checkPOS(word)] ## 리스트붕괴
sentences3 = sentences2.apply(lambda x: [word[0] for word in x if checkPOS(word)])
sentences3
```

###Gender bias 측정
참고 페이지 : https://chanind.github.io/nlp/2021/06/10/word2vec-gender-bias.html

1) gender vector를 구한다. 아래는 구하는 과정

```python
import numpy as np

#확실하게 남성과 여성의 차이를 나타내는 단어
gender_pairs=[
    ('남자','여자'),
    ('남','여'),
    ('아들','딸'),
    ('형','누나'),
    ('남동생','여동생'),
    ('아빠','엄마'),
    ('아버지','어머니'),
    #('남자아이','여자아이'), 단어 없음
    #('남자친구','여자친구'), 단어 없음
    ('할아버지','할머니'),
    ('외삼촌','이모'),
    ('삼촌','고모')
]
#남/여 벡터간 차이 계산 후 평균 => gender벡터 
gender_vectors=[]

for pair in gender_pairs :
    model.save('w2v_model')
    saved_model = Word2Vec.load('w2v_model')
    gender_vectors.append(model.wv[pair[0]]-model.wv[pair[1]])

gender_vector = np.mean(gender_vectors, axis=0)
```

2) gender 벡터에 확실하게 남자인 단어, 여자인 단어 벡터 정사영(projection)을 할 것. 예시 코드

```python
import numpy as np

queen_gender_projection = np.dot(model.wv["여왕"], gender_vector)
print(queen_gender_projection)
```

각 남자와 여자 그룹의 정사영 평균을 구한다

```python
male_projections = [np.dot(model.wv[word[0]], gender_vector) for word in gender_pairs]
female_projections = [np.dot(model.wv[word[1]], gender_vector) for word in gender_pairs]

mean_male_projection = np.mean(male_projections)
mean_female_projection = np.mean(female_projections)
print(mean_male_projection,mean_female_projection)
```

3) 이후, 임의의 단어를 정사영 시켰을 때 남자그룹의 정사영값, 여자그룹의 정사영 값중 어디에 더 가까운지 비교

```python
#예시 코드
test_word = "간호사"
test_word_projection = np.dot(model.wv[test_word], gender_vector)

mean_projection = (mean_male_projection + mean_female_projection) / 2
# scale the score so > 0 means female bias, < 0 means male bias : 결과값이 0보다 크면 female bias, 작으면 mail bias, 절대값이 클수록 bias가 더 심함
test_word_score = 2 * (test_word_projection - mean_projection) / (mean_female_projection - mean_male_projection)

print(test_word_score)
```

```python
male_bias=[]#남성에게 bias된 단어
female_bias=[]#여성에게 bias된 단어
avoid_overlap=[]#중복 피하기 위한 집합

for outer_array in sentences4: #이중 배열 접근
    for word in outer_array :#이중배열 안의 단어 접근
        if word not in avoid_overlap:
            test_word=word
            avoid_overlap.append(test_word)
            try : #없는 단어 예외처리
                test_word_projection = np.dot(model.wv[test_word], gender_vector)
                mean_projection = (mean_male_projection + mean_female_projection) / 2
                test_word_score = 2 * (test_word_projection - mean_projection) / (mean_female_projection - mean_male_projection)
            except KeyError:
                pass
            if test_word_score>0 :
                female_bias.append([test_word_score,word])
            elif test_word_score<0 :
                male_bias.append([test_word_score,word])
```

```python
male_bias.sort(key=lambda x:x[0])
female_bias.sort(key=lambda x:x[0])
female_bias.reverse()

print(male_bias[:19])
print(female_bias[:19])
```
